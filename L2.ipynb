##### 2) Напишите калькулятор который запрашивает на входе две переменные и знак, и в соответствии с знаком ( + - * / ) выводит результат

```python
a = int(input("Введите первый множитель:"))
b = int(input("Введите второй множитель:"))
c = input("Введите знак:")

match c:
    case "+":
        print(a + b)
    case "-":
        print(a-b)
    case "*":
        print(a*b)
    case "/":
        print(a/b)
    case _:
        print("не понял")
```
